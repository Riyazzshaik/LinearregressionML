```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
import matplotlib.pyplot as pyplot
import pickle
from matplotlib import style
```

```python
data=pd.read_csv("student-mat.csv",sep=";")
data
```

```python
data.head()
```

```python
data=data[["G1","G2","G3","studytime","failures","absences"]]
data
```

```python
predict ="G3"

X=np.array(data.drop([predict], axis=1))
y=np.array(data[predict])

x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.1)

linear= linear_model.LinearRegression()

linear.fit(x_train,y_train)
acc=linear.score(x_test,y_test)
print(acc)
```

```python
print('Coefficent : \n',linear.coef_)
print('Intercept : \n ',linear.intercept_)
```

```python
pyplot.scatter(data["G1"],data["G3"])
pyplot.xlabel("G1")
pyplot.ylabel("G3")
```

```python
predictions=linear.predict(x_test)
for x in range(len(predictions)):
  print(predictions[x],x_test[x],y_test[x])
```

```python
pyplot.scatter(y_test, predictions)
pyplot.xlabel("Actual G3 Values")
pyplot.ylabel("Predicted G3 Values")
pyplot.title("Actual vs Predicted G3 Scores")
pyplot.show()
```

# Task
Perform a comprehensive residual analysis and visualize the linear regression model's performance by creating:
1. A scatter plot of residuals (actual G3 values minus predicted G3 values) against the predicted G3 values.
2. A histogram and KDE plot of the residuals to check for normality.
3. A scatter plot of actual G3 values versus predicted G3 values, including a red diagonal line representing perfect predictions (y=x).

## Calculate Residuals

### Subtask:
First, we need to calculate the residuals of the model. Residuals are the difference between the actual observed values (y_test) and the values predicted by our model (predictions). These will be crucial for the subsequent visualization steps.


**Reasoning**:
I need to calculate the residuals by subtracting the predicted values from the actual test values, as per the instructions.



```python
residuals = y_test - predictions
print("Residuals calculated successfully:")
print(residuals)
```

**Reasoning**:
Now that the residuals are calculated, the next step is to create a scatter plot of residuals against the predicted G3 values as specified in the main task. This plot helps to visualize the spread of errors.



```python
pyplot.figure(figsize=(10, 6))
pyplot.scatter(predictions, residuals)
pyplot.axhline(y=0, color='r', linestyle='--')
pyplot.xlabel("Predicted G3 Values")
pyplot.ylabel("Residuals (Actual - Predicted)")
pyplot.title("Residuals vs Predicted G3 Values")
pyplot.grid(True)
pyplot.show()
```

**Reasoning**:
To assess the normality of the residuals, I need to create a histogram and a Kernel Density Estimate (KDE) plot of the calculated residuals, as specified in the main task.



```python
pyplot.figure(figsize=(10, 6))
sns.histplot(residuals,kde=True, bins=10)
pyplot.xlabel("Residuals")
pyplot.ylabel("Frequency")
pyplot.title("Histogram and KDE of Residuals")
pyplot.grid(True)
pyplot.show()
```

**Reasoning**:
The final step in the main task is to create a scatter plot of actual G3 values versus predicted G3 values, with a diagonal line indicating perfect predictions. This will visually represent the model's overall performance.



```python
pyplot.axes(projection="3d")
pyplot.figure(figsize=(10, 6))
pyplot.scatter(y_test, predictions)
pyplot.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', lw=2)
pyplot.xlabel("Actual G3 Values")
pyplot.ylabel("Predicted G3 Values")
pyplot.zlabel("Residuals")
pyplot.title("Actual vs Predicted G3 Values")
pyplot.grid(True)
pyplot.show()
```

```python
from scipy import stats

# Shapiro-Wilk Test
shapiro_test_statistic, shapiro_p_value = stats.shapiro(residuals)
print(f"Shapiro-Wilk Test: statistic={shapiro_test_statistic:.3f}, p-value={shapiro_p_value:.3f}")

# D'Agostino's K^2 Test
k2_test_statistic, k2_p_value = stats.normaltest(residuals)
print(f"D'Agostino's K^2 Test: statistic={k2_test_statistic:.3f}, p-value={k2_p_value:.3f}")

# Interpretation
alpha = 0.05
print("\nInterpretation:")
if shapiro_p_value > alpha and k2_p_value > alpha:
    print("Both tests suggest that the residuals look Gaussian (fail to reject H0).")
elif shapiro_p_value > alpha and k2_p_value <= alpha:
    print("Shapiro-Wilk suggests normality, but D'Agostino's K^2 does not. Consider larger sample size or other tests.")
elif shapiro_p_value <= alpha and k2_p_value > alpha:
    print("D'Agostino's K^2 suggests normality, but Shapiro-Wilk does not. This might indicate issues for smaller samples.")
else:
    print("Both tests suggest that the residuals are NOT Gaussian (reject H0).")
```
